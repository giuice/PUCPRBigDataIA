```python
#df.groupby('bairro')[['preco']].mean()
bairros = df['bairro'].unique()
fig = plt.figure(figsize=(18, 16))
for b in range(0,len(bairros)):
    bairro = bairros[b]
    ax = fig.add_subplot(4,2,b+1)
    ax.set_title(bairro)
    ax.hist(df.loc[df['bairro'] == bairro]['preco'])
    ax.tick_params(bottom="off", top="off", left="off", right="off")
    for key,spine in ax.spines.items():
        spine.set_visible(False)
plt.show()
   
```
